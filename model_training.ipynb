```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('datasets/Disease and symptoms dataset.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
encoder = LabelEncoder()
```

```python
df.diseases = encoder.fit_transform(df.diseases)
```

```python
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
len(df.columns)
```

```python
len(df.diseases.unique())
```

```python
X = df.drop('diseases', axis=1)
y = df.diseases
```

```python
X.shape
```

```python
model = keras.Sequential([
    keras.Input(shape=(377,)),
    keras.layers.Dense(512, activation='relu'),
    keras.layers.Dense(256, activation='relu'),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dense(773, activation='softmax')
])
```

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(X, y, epochs=40, batch_size=256)
```

```python
loss,accuracy = model.evaluate(X, y)
accuracy
```

```python
model.save("model/disease_model.keras")
```
